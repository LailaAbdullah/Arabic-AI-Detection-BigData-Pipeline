```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.appName('EDA').master('local[*]').getOrCreate()
spark.sparkContext.setLogLevel('WARN')
print('Spark version:', spark.version)
```

```python
sdf = spark.read.parquet('hdfs://localhost:9000/user/arabic_ai_detection/data/raw/raw_data.parquet')
print('Total rows:', sdf.count())
sdf.printSchema()
```

```python
# Label Distribution
sdf.groupBy('label').count().show()
```

```python
# Text Length Stats
sdf.withColumn('length', F.length('text')).groupBy('label').agg(
    F.round(F.avg('length'),2).alias('avg_length'),
    F.min('length').alias('min_length'),
    F.max('length').alias('max_length')
).show()
```

```python
# Missing values
sdf.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in sdf.columns]).show()
spark.stop()
```
